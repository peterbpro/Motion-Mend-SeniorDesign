```python

### Edit to get the path to the 'Simulation' folder ####
FILEPATH = 'C:\\Users\\kenzo\\Documents\\SENIOR DESIGN\\MotionMend\\Hardware\\Experimental\\Simulation'
### Edit to get the path to the 'Simulation' folder ####

modelFileName = os.path.join(FILEPATH, 'models','Rajagopal_2015.osim')          # The path to an input model
outputCalibratedFileName = os.path.join(FILEPATH, 'models', 'calibrated_model.osim')
orientationsFileName = os.path.join(FILEPATH, 'data', 'data_2024_11_06_17_43.sto')   # The path to orientation data for calibration 
resultsDirectory = os.path.join(FILEPATH, 'IK_results')
motionFileName = os.path.join(resultsDirectory, 'ik_data_2024_11_07_19_01.mot')
sensor_to_opensim_rotation = osim.Vec3(-pi/2, 0, 0)

# Load and Initialize the model
model = osim.Model(modelFileName)
state = model.initSystem()

#Load Motion
q = osim.TimeSeriesTable(motionFileName)
time = q.getIndependentColumn()

relevant_coords = set([
    'hip_flexion_r',
    'hip_adduction_r',
    'hip_rotation_r',
    'knee_angle_r',
    'knee_angle_r_beta',
    'hip_flexion_l',
    'hip_adduction_l',
    'hip_rotation_l',
    'knee_angle_l',
    'knee_angle_l_beta'
])
    
```

```python


for i,t in enumerate(time):
    value=osim.RowVector(q.getRowAtIndex(i))
    for j, coordinate in enumerate(model.updCoordinateSet()):
        if coordinate.getName() in relevant_coords:
            print(t, coordinate.getName(), value[j])
```

```python
imuIK = osim.IMUInverseKinematicsTool()
# Set tool properties
imuIK.set_model_file(outputCalibratedFileName)
imuIK.set_orientations_file('C:\\Users\\kenzo\\Documents\\SENIOR DESIGN\\MotionMend\\Hardware\\Experimental\\Simulation\\data\\ik_data_2024_11_07_19_01.mot')
imuIK.set_sensor_to_opensim_rotations(sensor_to_opensim_rotation)
imuIK.set_results_directory(resultsDirectory)

# Set time range in seconds
imuIK.set_time_range(0, 0) 
imuIK.set_time_range(1, 1000)
 
```

```python
imuIK.run()
```
